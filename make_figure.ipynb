```python
import matplotlib.pyplot as plt

# Define a function to parse the accuracy data from a file
def parse_accuracy_file(filename):
    with open(filename, 'r') as file:
        # Use an iterator instead of a list
        lines = iter(file.readlines())
    
    # Initialize empty lists to store model numbers and accuracy values
    models = []
    accuracies_list = []
    
    # Skip the header line
    next(lines)
    
    # Iterate through each line, extracting model numbers and accuracy values
    for line in lines:
        if not line.startswith('Model:'):  # Skip lines that do not start with 'Model:'
            continue
        # The rest of the code remains unchanged
        model_name = line.split(': ')[1].strip().split('.dat')[0]  # Extract model name
        accuracy = line.split(', ')[1].strip().split(': ')[1].strip()  # Extract accuracy
        models.append(int(model_name.replace('model_', '')))
        accuracies_list.append(float(accuracy))
    
    return models, accuracies_list

# Parse the accuracy file
models, accuracies = parse_accuracy_file('accuracies.txt')
print(models)
print(accuracies)

```

```python
# Use matplotlib to plot the chart
plt.figure(figsize=(12, 8), dpi=100)  # Set the figure size and DPI for better resolution
plt.plot(models, accuracies, marker='o', linestyle='-', color='orange')  # Plot the line graph
plt.title('Model Accuracy Chart')  # Set the title of the chart
plt.xlabel('Percentage of Gutenberg dataset used for training')  # Set the label for the x-axis
plt.ylabel('Predict accuracy')  # Set the label for the y-axis
plt.savefig('model_accuracy_chart.png', dpi=300, bbox_inches='tight')# Save the plot to a file
plt.show()  # Display the chart
```
